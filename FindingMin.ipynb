### Find y'=0 in a stupid way

```python
h = 0.001
x = np.arange(-2, 2, h)
```

```python
x
```

```python
y = np.sin(x**2)
```

```python
y
```

Plotting y=x

```python
plt.plot(x,y)
```

Calculating the derivative

```python
yprime = np.gradient(y, h)
```

```python
yprime
```

```python
plt.plot(x,yprime)
plt.plot(x, np.repeat(0, len(x)))
```

Finding the values where y'=0

Start by finding the smallest values of y'

```python
np.min(abs(yprime))
```

Only in x = 0 (-0.00000000000022) is |y'| less then 0.000001

```python
x[np.abs(yprime)<10e-6]
```

```python
x[np.abs(yprime)<12e-3]
```
